```python
col_use=['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance',
        'CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
df=dataset[col_use]
df
```

```python
df.isnull().sum() #OR dataset.isna().sum()
```

```python
df.columns
```

```python
df['Car']=df.Car.fillna(df.Car.median())
df['BuildingArea']=df.BuildingArea.fillna(df.BuildingArea.median())
df
```

```python
df.isnull().sum()
```

```python
df.fillna(method='pad')
```

```python
df.isnull().sum()
```

```python
df['CouncilArea'].value_counts()
```

```python
newdf=df.fillna({'CouncilArea':'Macedon Ranges'})
newdf
```

```python
newdf.isnull().sum()
```

```python
newdf.shape
```

```python
newdf=pd.get_dummies(newdf,drop_first=True)
newdf
```

```python
x=newdf.drop(['Price'],axis=1)
y=newdf[['Price']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression().fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

# Lasso(L1)Regression

```python
from sklearn import linear_model
lasso_reg=linear_model.Lasso(alpha=0.1,max_iter=1000,tol=0.1)
lasso_reg.fit(x_train,y_train)
```

```python
lasso_reg.score(x_train,y_train)
```

```python
lasso_reg.score(x_test,y_test)
```

# Ridge(L2)Regression

```python
from sklearn.linear_model import Ridge
ridge_reg=Ridge(alpha=0.1,max_iter=1000,tol=0.1)
ridge_reg.fit(x_train,y_train)
```

```python
ridge_reg.score(x_train,y_train)
```

```python
ridge_reg.score(x_test,y_test)
```
